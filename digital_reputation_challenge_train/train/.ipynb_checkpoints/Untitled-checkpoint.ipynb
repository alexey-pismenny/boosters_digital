```python
import pandas as pd
import numpy as np
```

```python
x3_pd = pd.read_csv('X3.csv')
y_pd = pd.read_csv('Y.csv')
#x1_pd['y1'] = y_pd['1']
#x1_pd['y2'] = y_pd['2']
#x1_pd['y3'] = y_pd['3']
#x1_pd['y4'] = y_pd['4']
#x1_pd['y5'] = y_pd['5']
```

```python
X_train = np.array(x3_pd[[str(i) for i in range(1, 453)]])
```

```python
x3_pd_test = pd.read_csv('X3_test.csv')
```

```python
x3_pd_test
```

```python
X_test = np.array(x3_pd_test[[str(i) for i in range(1, 453)]])
```

```python
X_test
```

```python
Y_train = np.array(y_pd[['1', '2', '3', '4', '5']])
```

```python
Y_train
```

```python
with open('3.csv', 'w') as f:
    for i in range(X_test.shape[0]):
        res = np.array([0., 0., 0., 0., 0.])
        w_sum = 0.
        for j in range(X_train.shape[0]):
            w = len((X_train[j] * X_test[i]).nonzero()[0])
            res += w * Y_train[j]
            w_sum += w
        res /= w_sum
        res = [str(val) for val in res]
        f.write(x3_pd_test['id'][i] + ',' + ','.join(res.tolist()) + '\n')
```
